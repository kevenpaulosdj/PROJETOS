# Anlise de Dados - Indicadores educacionais do Estado da Bahia



# CONTEXTUALIZAO E OBJETIVOS
O projeto a seguir consiste em uma anlise exploratria de quatro etapas (observao, anlise, formulao e validao) a respeito dos indicadores educacionais do Estado da Bahia. Os objetivos desse projeto esto focados no tratamento, na coleta e na observao inicial dos dados extrados do site do Qedu, que englobam fatores como Ideb, Saeb, taxas de distoro e rendimento e permanncias. A partir disso, foca-se na obteno de insights valiosos a partir desses dados - representados atravs de tabelas ou grficos - de modo que eles sejam capazes de contribuir na resoluo de problemas e na proposio de eventuais solues.

# PRINCIPAIS INSIGHTS
O percentual de alunos matriculados na gerao  superior a 90%, aumentando conforme o passar dos anos. J os valores do Ideb (ndice de Desenvolvimento da Educao Bsica) e da taxa de aprendizado variam de acordo com fatores como ciclo (estgio escolar em que o aluno se encontra) e dependncia (referente ao ambiente de estudo do aluno). Nos casos analisados, respectivamente, nota-se que alunos que encontram-se nos anos iniciais do Ensino Fundamental e/ou estudam em escolas privadas tendem a apresentar melhores ndices desses dois indicadores, com certa vantagem para alunos que estudam em escolas da rede pblica e/ou esto no Ensino Mdio. Quanto s notas mdias das disciplinas Matemtica e Lngua Portuguesa, percebe-se que estudantes do Ensino Mdio possuem maiores notas em relao a alunos dos demais ciclos.

No que tange ao percentual de domnio de ambas as matrias, entretanto, tal situao inverte-se: alunos do Ensino Fundamental possuem maior conhecimento em relao aos estudantes do segundo grau. Alm disso, discentes que estudam em escolas das redes pblica, municipal ou estadual apresentam menor domnio - seja ele bsico ou adequado - das matrias citadas anteriormente em relao queles que estudam em escolas federais ou privadas. Essa tendncia reflete em outros indicadores, como o de rendimento, haja vista que o baixo saber dos conceitos bsicos de cada disciplina pode ser considerado como um dos fatores decisivos para o aumento ou diminuio dos percentuais de reprovao e abandono, relativamente maiores em instituies de ensino financiadas pelo Estado. O mesmo cenrio se repete em relao  taxa de distoro idade-srie, que nada mais  que o percentual de alunos que esto matriculados em uma srie, mas tm uma idade superior  recomendada para tal.

Diante do exposto, pode-se inferir que alunos de escolas privadas ou federais nas cidades geralmente tm um ambiente mais favorvel para aprender em relao aos demais, visto que as instituies de ensino citadas costumam ter melhor infraestrutura, mais materiais didticos e professores bem preparados. Alm disso, esses alunos, em muitos casos, vm de famlias com maior estabilidade financeira, o que ajuda a garantir um acompanhamento mais prximo dos estudos. O acesso  tecnologia tambm faz diferena, facilitando o aprendizado. Por outro lado, a falta de recursos, a longa distncia at a escola, a precariedade do transporte e at mesmo a necessidade de muitos alunos ajudarem em casa acabam dificultando a continuidade nos estudos de estudantes matriculados em escolas estaduais, pblicas ou municipais. Com menos suporte e condies adversas, muitos acabam desistindo ou enfrentando dificuldades para acompanhar o ritmo das aulas.

Em relao aos ciclos de ensino, nota-se que no Ensino Mdio  quando os desafios comeam de verdade. As matrias ficam mais difceis, a cobrana cresce e muitos jovens precisam dividir o tempo entre a escola, o trabalho e as responsabilidades em casa. Alm disso, a presso para decidir o futuro pode ser grande, e nem todos tm o apoio necessrio para lidar com isso. Sem estrutura e suporte adequados, fica mais fcil se perder no caminho, o que explica os altos ndices de reprovao, abandono e atraso escolar. Isso mostra que, conforme os alunos avanam de srie, a escola se torna mais exigente, e aqueles que no recebem o acompanhamento necessrio acabam enfrentando mais dificuldades para seguir em frente.
