# Análise de Dados - Indicadores educacionais do Estado da Bahia



# CONTEXTUALIZAÇÃO E OBJETIVOS
O projeto a seguir consiste em uma análise exploratória de quatro etapas (observação, análise, formulação e validação) a respeito dos indicadores educacionais do Estado da Bahia. Os objetivos desse projeto estão focados no tratamento, na coleta e na observação inicial dos dados extraídos do site do Qedu, que englobam fatores como Ideb, Saeb, taxas de distorção e rendimento e permanências. A partir disso, foca-se na obtenção de insights valiosos a partir desses dados - representados através de tabelas ou gráficos - de modo que eles sejam capazes de contribuir na resolução de problemas e na proposição de eventuais soluções.

# TECNOLOGIAS UTILIZADAS
A linguagem de programação Python foi a utilizada durante as quatro etapas do projeto realizado, representada por bibliotecas importadas, como o Pandas, o Matplotlib e o Seaborn. Enquanto a primeira biblioteca foi bastante utilizada na primeira etapa, sendo responsável por ajudar de maneira efetiva na manipulação de tabelas, as duas últimas foram utilizadas na etapa seguinte para plotagem e manipulação de gráficos, exercendo de maneira conjunta com os conceitos básicos da Estatística um papel fundamental na visualização dos dados. Já nas duas últimas etapas, o uso do raciocínio lógico e do pensamento analítico foram bastante primordiais para o andamento e a conclusão do projeto como um todo, visto que estes foram responsáveis por gerar, justificar e confirmar pontos de vista, um processo que ocorre de maneira rotineira no cotidiano de um analista de dados.

# DEPENDÊNCIAS NECESSÁRIAS

- Quais são as taxas de aprovação e reprovação dos alunos baianos?
- Quais são os respectivos percentuais dos alunos baianos que possuem dificuldade em Matemática ou em Língua Portuguesa?

# PRINCIPAIS INSIGHTS
O percentual de alunos matriculados na geração é superior a 90%, aumentando conforme o passar dos anos. Já os valores do Ideb (Índice de Desenvolvimento da Educação Básica) e da taxa de aprendizado variam de acordo com fatores como ciclo (estágio escolar em que o aluno se encontra) e dependência (referente ao ambiente de estudo do aluno). Nos casos analisados, respectivamente, nota-se que alunos que encontram-se nos anos iniciais do Ensino Fundamental e/ou estudam em escolas privadas tendem a apresentar melhores índices desses dois indicadores, com certa vantagem para alunos que estudam em escolas da rede pública e/ou estão no Ensino Médio. Quanto às notas médias das disciplinas Matemática e Língua Portuguesa, percebe-se que estudantes do Ensino Médio possuem maiores notas em relação a alunos dos demais ciclos.

No que tange ao percentual de domínio de ambas as matérias, entretanto, tal situação inverte-se: alunos do Ensino Fundamental possuem maior conhecimento em relação aos estudantes do segundo grau. Além disso, discentes que estudam em escolas das redes pública, municipal ou estadual apresentam menor domínio - seja ele básico ou adequado - das matérias citadas anteriormente em relação àqueles que estudam em escolas federais ou privadas. Essa tendência reflete em outros indicadores, como o de rendimento, haja vista que o baixo saber dos conceitos básicos de cada disciplina pode ser considerado como um dos fatores decisivos para o aumento ou diminuição dos percentuais de reprovação e abandono, relativamente maiores em instituições de ensino financiadas pelo Estado. O mesmo cenário se repete em relação à taxa de distorção idade-série, que nada mais é que o percentual de alunos que estão matriculados em uma série, mas têm uma idade superior à recomendada para tal.

Diante do exposto, pode-se inferir que alunos de escolas privadas ou federais nas cidades geralmente têm um ambiente mais favorável para aprender em relação aos demais, visto que as instituições de ensino citadas costumam ter melhor infraestrutura, mais materiais didáticos e professores bem preparados. Além disso, esses alunos, em muitos casos, vêm de famílias com maior estabilidade financeira, o que ajuda a garantir um acompanhamento mais próximo dos estudos. O acesso à tecnologia também faz diferença, facilitando o aprendizado. Por outro lado, a falta de recursos, a longa distância até a escola, a precariedade do transporte e até mesmo a necessidade de muitos alunos ajudarem em casa acabam dificultando a continuidade nos estudos de estudantes matriculados em escolas estaduais, públicas ou municipais. Com menos suporte e condições adversas, muitos acabam desistindo ou enfrentando dificuldades para acompanhar o ritmo das aulas.

Em relação aos ciclos de ensino, nota-se que no Ensino Médio é quando os desafios começam de verdade. As matérias ficam mais difíceis, a cobrança cresce e muitos jovens precisam dividir o tempo entre a escola, o trabalho e as responsabilidades em casa. Além disso, a pressão para decidir o futuro pode ser grande, e nem todos têm o apoio necessário para lidar com isso. Sem estrutura e suporte adequados, fica mais fácil se perder no caminho, o que explica os altos índices de reprovação, abandono e atraso escolar. Isso mostra que, conforme os alunos avançam de série, a escola se torna mais exigente, e aqueles que não recebem o acompanhamento necessário acabam enfrentando mais dificuldades para seguir em frente.

### ALGUMAS IMAGENS









# PRÓXIMOS PASSOS

Faz-se necessária a realização de investimentos direcionados à rede pública de educação por parte do Governo do Estado, representado pela figura da Secretaria da Educação (SEC), de modo a mitigar as desigualdades presentes no ambiente escolar baiano e democratizar o acesso ao ensino. Cabe também à Secretaria da Educação, em ação conjunta com a Secretaria de Infraestrutura da Bahia (Seinfra), a realização de ações que tenham o objetivo de conectar os estudantes aos seus respectivos locais de ensino. Tais ações podem envolver a inclusão digital, através da concessão de smartphones e tablets com acesso à internet aos alunos, bem como a construção de novas escolas e circulação de transportes escolares em locais remotos.

Além disso, o Estado, representado pelas figuras da Secretaria da Assistência e Desenvolvimento Social (SEADES) e da Secretaria do Trabalho, Emprego, Renda e Esporte (SETRE), deve pôr em prática medidas que visem amenizar os impactos gerados pela desigualdade social no país, como a criação de novos programas sociais que visem beneficiar famílias de baixa renda, bem como a geração de novos empregos e a melhora nas condições de trabalho dos já existentes. Outrossim, para que as medidas anteriores operem perfeitamente sem que haja alteração na rotina das crianças e jovens, os órgãos do Governo do Estado devem fiscalizar e proibir quaisquer maneiras de trabalho infantojuvenil, salvo àqueles que possuem representação legal por lei, a exemplo do Jovem Aprendiz.

Tendo em vista que a saúde mental dos estudantes, sobretudo adolescentes, é algo que comprovadamente pode afetar o desempenho dos mesmos em caso de não haver o devido cuidado, devem-se contratar profissionais especializados na área da Psicologia para os ambientes de ensino, de modo que estes fiquem responsáveis por acompanhar a rotina dos jovens e aconselhá-los em eventuais dificuldades encontradas. Por fim, mas não menos importante, deve-se priorizar o aprendizado dos conceitos básicos de cada matéria em detrimento da aprovação dos alunos, sobretudo em séries menores. Dessa forma, garante-se que os estudantes estarão realmente aprendendo e não "decorando para passar de ano".
